# **Read the heading and paragraph text of article using beautifulsoup and store into new column extracted text.**  

```python
import regex
```

```python
text_list=[]
for i in range(len(df['URL'])):
    url=df['URL'][i]
    res=requests.get(url)
    soup=BeautifulSoup(res.text,'html.parser')
    text_content=''
    text_para=''
    text_heading=''
    for tag in soup.find_all('h1'): 
      text_heading=tag.text
    for para in soup.find_all('p',class_=""):
      text_para=text_para+' '+para.text

    total_text=text_heading+':'+text_para
    text_list.append(total_text)
```

```python
df['extracted_text']=text_list
```

# **Read all the stopwords from stopwords directory.**

```python
import os
stop_word_list=[]
folder_path='/home/shiva1974/Desktop/text-web-scrapper-and-analizer/StopWords'
list_of_stopwordfiles=os.listdir('/home/shiva1974/Desktop/text-web-scrapper-and-analizer/StopWords')
for i in list_of_stopwordfiles:
    full_path=folder_path+'/'+i
    stop_words=open(full_path,'rb')
    stop_words=[word for word in stop_words.readlines()]
    stop_word_list.extend(stop_words)
```

```python
for i in range(len(stop_word_list)):
    stop_word_list[i]=str(stop_word_list[i])
```

```python
stop_word_list
```

```python
import regex
for i in range(len(stop_word_list)):
    stop_word_list[i]=regex.sub(r'[S\\]','',stop_word_list[i])
    stop_word_list[i]=regex.sub(r'[\\rr\\S\\]','r',stop_word_list[i])
    stop_word_list[i]=regex.sub(r'[\\n]','n',stop_word_list[i])
    stop_word_list[i]=regex.sub(r'[b'']','',stop_word_list[i])
    stop_word_list[i]=stop_word_list[i][1:len(stop_word_list[i])]
    stop_word_list[i]=stop_word_list[i][0:len(stop_word_list[i])-1]
```

```python
stop_word_list
```

```python
for i in range(len(stop_word_list)):
    stop_word_list[i]=stop_word_list[i][0:len(stop_word_list[i])-2]
```

```python
new_stop_word_list=[]
for i in stop_word_list:
    i=i.split('|')
    new_stop_word_list.extend(i)
```

# **Count the positive and negative words and store in dataframe’s new columns.**

```python
extracted_positives=[]
for i in df['extracted_text']:
    extracted_positives_from_url=[]
    for j in i:
        if j in new_positive_words:
            extracted_positives_from_url.append(j)
    extracted_positives.append(extracted_positives_from_url)

extracted_negatives=[]
for i in df['extracted_text']:
    extracted_negatives_from_url=[]
    for j in i:
        if j in new_negative_words:
            extracted_negatives_from_url.append(j)
    extracted_negatives.append(extracted_negatives_from_url)

for i in range(len(extracted_positives)):
    extracted_positives[i]=len(extracted_positives[i])
for i in range(len(extracted_negatives)):
    extracted_negatives[i]=len(extracted_negatives[i])
df['POSITIVE SCORE']=extracted_positives
df['NEGATIVE SCORE']=extracted_negatives
```

# **Using The positive words and Negative words calculate the polarity score and store them in dataframe too.**

```python
calculated_polarity=[]
for i in df.index:
    value=(df['POSITIVE SCORE'][i]-df['NEGATIVE SCORE'][i])/( (df['POSITIVE SCORE'][i]+df['NEGATIVE SCORE'][i])+0.000001 )
    calculated_polarity.append(value)
df['POLARITY SCORE']=calculated_polarity
```

# **Using ‘nltk .corpus.cmudict’  find the syllables and then find the percentage of complex word.**

```python
from nltk.corpus import cmudict
import string
nltk.download('cmudict')

phoneme_dict = dict(cmudict.entries())

def syllables_in_word(word):
        if word in phoneme_dict:   
            return len( [ph for ph in phoneme_dict[word] if ph.strip(string.ascii_letters)] )   # more destructive; less efficient? NO! see timeit results in my comments below
        else:        
            return 0  
percentage_of_complex=[]
for mylist in df['extracted_text']:
    count = 0

    for myword in mylist:
        value=syllables_in_word(myword)
        if value>2:
             count+=1         
        
    percentage=(count/len(mylist))*100
    percentage_of_complex.append(percentage)
df['PERCENTAGE OF COMPLEX WORD']=percentage_of_complex
```

# **calculate word count per article excluding stop words.**

```python
word_count_per_row=[]
for i in df.index:
    row=df['extracted_text'][i]
    for j in range(len(row)):
        row[j]=regex.sub(r'[?!,.]','',row[j])
    word_count_per_row.append(len(row))
df['WORD COUNT']=word_count_per_row
```

# **Using pre written syllable function calculate syllable per word value.**

```python
syllables_per_word=[]
for i in df.index:
    syllables=0
    for myword in df['extracted_text'][i]:
        syllables+=syllables_in_word(myword)
    value=syllables/len(df['extracted_text'][i])
    syllables_per_word.append(value)
df['SYLLABLE PER WORD']=syllables_per_word
```

# **Using regex calculate personal pronouns and rember to exclude ‘US’ word.**

```python
import regex
pronounRegex=regex.compile(r'\b(I|we|my|ours|(?-i:us))\b',regex.I)
pronouns=[]
for j in df1['extracted_text']:
    curr_pronouns=pronounRegex.findall(j)
    length=len(curr_pronouns)
    for i in range(length):
        if curr_pronouns[i] in ['US']:
            curr_pronouns.pop(curr_pronouns[i])
            length-=1  
    pronouns.append(len(curr_pronouns))
df['PERSONAL PRONOUNS']=pronouns
```
